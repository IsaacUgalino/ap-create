```python
import random

stocks = {
    "Johnson&Johnson": "Low",
    "Campbell Soup Company": "Low",
    "Visa": "Low",
    "Costco": "Somewhat",
    "Apple": "Somewhat",
    "Microsoft": "Somewhat",
    "Spotify": "Medium",
    "NVIDIA": "Medium",
    "Tesla": "Medium",
    "Snap Inc.": "High",
    "Palantir Technologies": "High",
    "Roblox Corp": "High",
    "GameStop": "Ultra High",
    "MicroStrategy": "Ultra High",
    "Oshkosh Corporation": "Ultra High",
    "Chipotle": "Special",
    "NVR": "Special",
    "Berkshire Hathaway": "Special"
}

stocks_prices = {
  'Johnson&Johnson': 154.58,
  'Campbell Soup Company': 36.29,
  'Visa': 278.49,
  'Costco': 748.50,
  'Apple': 209.28,
  'Microsoft': 391.85,
  'Spotify': 297.22,
  'NVIDIA': 111.01,
  'Tesla': 284.95,
  'Snap Inc.': 14.82,
  'Palantir Technologies': 112.78,
  'Roblox Corp': 40.09,
  'GameStop': 27.46,
  'MicroStrategy': 368.71,
  'Oshkosh Corporation': 104.55,
  'Chipotle': 2913.48,
  'NVR': 8098.20,
  'Berkshire Hathaway': 795760.00,
}

volatility_ranges = {
  "Low": (-0.05, 0.05),
  "Somewhat": (-0.10, 0.10),
  "Medium": (-0.20, 0.20),
  "High": (-0.30, 0.30),
  "Ultra High": (-0.45, 0.45),
  "Special": (-0.90, 0.90)
}

investments = {
  "Johnson&Johnson": 0,
  "Campbell Soup Company": 0,
  "Visa": 0,
  "Costco": 0,
  "Apple": 0,
  "Microsoft": 0,
  "Spotify": 0,
  "NVIDIA": 0,
  "Tesla": 0,
  "Snap Inc.": 0,
  "Palantir Technologies": 0,
  "Roblox Corp": 0,
  "GameStop": 0,
  "MicroStrategy": 0,
  "Oshkosh Corporation": 0,
  "Chipotle": 0,
  "NVR": 0,
  "Berkshire Hathaway": 0
}

def main():
  account_balance = 25000
  simulated_days = 0
  run = True

  while run == True:
    user_choice = menu(account_balance, simulated_days)
    if user_choice == '1':
      option = input("Would you like to buy or sell?: ").lower()
      if option == 'buy':
        account_balance = buy(investments, account_balance)
      elif option == 'sell':
        account_balance = sell(investments, account_balance)
    elif user_choice == '2':
      simulated_days = simulate(simulated_days)
    elif user_choice == '3':
      add_stocks()
    elif user_choice == '4':
      remove_stocks()
    elif user_choice == '5':
      view()
    elif user_choice == '6':
      confirm = input("Are you sure you want to quit? Yes/No: ").lower()
      if confirm == "yes":
        run = False
      else:
        pass
  print("Thank you for investing your time <3!")

def menu(account_balance, simulated_days):
    print(f"Day: {simulated_days}")
    print(f'Account Value: {account_balance}')
    print("1. $$$INVEST$$$")
    print("2. Simulate Day")
    print("3. Add Company")
    print("4. Remove Company")
    print("5. View Companies")
    print("6. Quit")
    choice = input("What would you like to do?: ")
    return choice

def buy(investments, account_balance): #Buy stocks
  print("_______________________________-BUYING-_________________________________")
  company = input("Which company would you like to invest in?: ").title()
  try:
    quantity = int(input("How much stock do you want to buy?: "))
    if company in stocks_prices.keys(): #Goes through dictionary to find the stock you want
      total = stocks_prices[company] * quantity
      if account_balance > total:
        account_balance -= total
        investments[company] = investments.get(company, 0) + quantity
        print(f"Total: ${total}")
        print(f'Congrats! You bought {quantity} shares of {company}.')
        print("_______________________________-BOUGHT-_________________________________")
        return account_balance
      else:
        print(f"Total: ${total}")
        print("You don't have enough money to make this purchase.")
        print(f"You need ${total - account_balance}")
        print("____________________________-LIMITREACHED-______________________________")
        return
    elif company != stocks_prices[company]:
      print("Invalid Company, use a name from the list.")
      print("_____________________________-KEYERROR-_________________________________")
    return
  except ValueError:
    print("Invalid Quantity, use whole number.")
    print("____________________________-VALUEERROR-________________________________")
    return

def sell(investments, account_balance):
  print("______________________________-SELL-____________________________________")
  for company, shares in investments.items():
    print(f'{company}: {shares} shares')
  try:
    company = input("Which company would you sell?: ").title()
  except ValueError:
    print("Invalid Quantity, use whole number.")
    print("____________________________-VALUEERROR-________________________________")
    return
  try:
    quantity = int(input("How much stock do you want to sell?: "))
    if company in stocks_prices.keys():
      returns = stocks_prices[company] * quantity
      if investments[company] >= quantity:
        account_balance += returns
        investments[company] = investments.get(company, 0) + quantity
        print(f"Return: ${returns}")
        print(f'Congrats! You sold {quantity} shares of {company}.')
        print("________________________________-SOLD-__________________________________")
        return account_balance
      else:
        print("You don't have enough shares to make this return.")
        print(f"You only have {investments[company]} shares")
        print("___________________________-LIMITREACHED-_______________________________")
      return
    elif str(company) != investments[company]:
      print("This company doesn't exist.")
      print("____________________________-NOTFOUND-__________________________________")
      return
  except KeyError:
    print("Invalid Company, use a name from the list.")
    print("______________________________-KEYERROR-________________________________")
    return

def simulate(simulated_days):
  print("_____________________________-SIMULATING-_______________________________")
  for company, vol in stocks.items():
      old_price = stocks_prices[company]
      change = round(random.uniform(*volatility_ranges[vol]), 2)
      gain_loss = round(1 + change, 2)
      stocks_prices[company] = round(stocks_prices[company] + (stocks_prices[company] * change), 2)
      if change > 0:
        print(f'{company}: UP {round(change *100, 2)}% | ${old_price} --> ${stocks_prices[company]}')
      elif change < 0:
        print(f'{company}: DOWN {round(change *100, 2)}% | ${old_price} --> ${stocks_prices[company]}')
      elif change == 0:
        print(f'{company}: NEUTRAL {round(change *100, 2)}%|${old_price} --> ${stocks_prices[company]}')
  simulated_days += 1
  print("____________________________-SIMULATED-_________________________________")
  return simulated_days

def add_stocks():
  print("_____________________________-ADDSTOCKS-________________________________")
  print("Low, Somewhat, Medium, High, Ultra High, Special")
  vol = input('Pick a volatility level: ').title()
  if vol == 'Low' or vol == 'Somewhat' or vol == 'Medium' or vol == 'High' or vol == 'Ultra High' or vol == 'Special':
    try:
      name = input("What would you like to name your company?: ").title()
      price = round(float(input(f'How much does each share of {name} cost?: ')), 2)
      stocks[name] = vol
      stocks_prices[name] = price
      investments[name] = 0
      print(f"{name} has been added to {vol} volatility at ${price} per share.")
      print("____________________________-STOCKADDED-________________________________")
      return
    except KeyError:
      print("Invalid Company, use a CATEGORY from the list.")
      print("______________________________-KEYERROR-________________________________")
      return
    else:
      print("Invalid volatility level.")
      print("__________________________-INVALIDVOLATILITY-___________________________")
      return

def remove_stocks():
  print("______________________________-DELETING-________________________________")
  name = input("Which company would you like to remove?: ").title()
  for company in stocks.keys():
    if name == company:
      del stocks[name]
      del stocks_prices[name]
      del investments[name]
      print(f"{name} successfully removed.")
      print("______________________________-DELETING-________________________________")
      return
  if name != stocks.values():
    print("Company not found. Try a different name.")
    print("______________________________-NOTFOUND-________________________________")

def view():
  print("________________________________-VIEW-__________________________________")
  print("1. Investments")
  print("2. Stock Prices")
  choice = input("Which would you like to view?: ")
  if choice == "1":
    print("_______________________________-VIEWING-________________________________")
    for company, quantity in investments.items():
      print(f'{company}: {quantity} Shares')
    print("_____________________________-INVESTMENTS-______________________________")
  elif choice == "2":
    print("_______________________________-VIEWING-________________________________")
    for company, prices in stocks_prices.items():
      print(f'{company}: ${prices}')
    print("___________________________-STOCKS&PRICES-______________________________")
  else:
    print("Invalid input. Please try 1 or 2.")
    print("____________________________-INVALIDINPUT-______________________________")

    return


if __name__ == "__main__":
  main()

```
